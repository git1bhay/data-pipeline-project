```python
from pyspark.sql.functions import col

VOLUME_PATH = "/Volumes/assignment_catalog/bronze_layer/assignment_csv_data/Assignment/*.csv"
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(VOLUME_PATH)


df = df.withColumn("source_file", col("_metadata.file_path"))

df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("assignment_catalog.bronze_layer.sales_raw")

df_table = spark.table("assignment_catalog.bronze_layer.sales_raw")
df_table.printSchema()
```
