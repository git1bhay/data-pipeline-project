```python
from pyspark.sql.functions import sum, col

df = spark.table("assignment_catalog.silver_layer.sales_clean")

df_gold = df.groupBy("date", "sku", "data_source").agg(
    sum("units").alias("total_units"),
    sum("revenue").alias("total_revenue")
)

# Rename column to match requirement
df_gold = df_gold.withColumnRenamed("sku", "product_identifier")

df_gold.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("assignment_catalog.gold_layer.daily_sales")


display(df_gold)
```

```python
df_gold.coalesce(1).write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv("/Volumes/assignment_catalog/gold_layer/output_volume/final_sales")
```
